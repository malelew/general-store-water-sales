```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy.stats as scs
import matplotlib.pyplot as plt
from bkcharts import Area, show, vplot, output_file, defaults
```

```python
df_15_16 = pd.read_csv('items-2015-09-21-2016-06-11.csv')
df_16_17 = pd.read_csv('items-2016-09-19-2017-06-13.csv')
```

```python
df_15_16 = df_15_16[df_15_16.Category == 'Drinks']
df_15_16 = df_15_16[df_15_16.Item == 'Water']
df_15_16
```

```python
df_16_17 = df_16_17[df_16_17.Category == 'Drinks']
df_16_17 = df_16_17[df_16_17.Item == 'Water']
df_16_17
```

```python
# clean out unneeded columns
columns_keep = ['Date', 'Item', 'Qty', 'Price Point Name', 'Gross Sales', 'Net Sales', 'Event Type']
df_16_17 = df_16_17[columns_keep]
df_15_16 = df_15_16[columns_keep]
```

```python
# rename the columns
column_names = ['date', 'item', 'qty', 'price_point_name', 'gross_sales', 'net_sales', 'event_type']
df_16_17.columns = column_names
df_15_16.columns = column_names
```

```python
# remove misc price point
df_15_16 = df_15_16[df_15_16.price_point_name != 'Boxed Just Water']
df_16_17 = df_16_17[df_16_17.price_point_name != 'Minis']
```

```python
# rename regular size since for the first few months the boxed water had not been renamed
df_16_17.price_point_name[df_16_17.price_point_name == 'Regular Size'] = 'Boxed Just Water'
```
